```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("data_generated-01.xlsx")
```

```python
df.days
```

```python
plt.title(label="Relation between Days and Loss-per-Bag")
plt.xlabel(xlabel="Days")
plt.ylabel(ylabel="Loss Per Bag")

plt.scatter(df.days,df['loss/bag'])
plt.savefig('fig1.png')
```

```python
plt.title(label="Relation between log(Days) and Loss-per-Bag")
plt.xlabel(xlabel="log(Days)")
plt.ylabel(ylabel="Loss Per Bag")

plt.scatter(df['log_days'],df['loss/bag'])
plt.savefig('fig1.png')
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error

X,y = df.days.to_numpy().reshape(-1, 1),df['loss/bag'].to_numpy().reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)
print("Intercept:", model.intercept_[0])
print("Slope:", model.coef_[0][0])
print("Correlation Cofficient", np.corrcoef(X.reshape(-1),y.reshape(-1))[0][1])
print("MAE",mean_absolute_error(y, model.predict(X)))
plt.scatter(X, y)
plt.plot(X, model.predict(X), 'r-')
plt.xlabel("Days")
plt.ylabel("Loss Per Bag")
plt.title("Relation between Days and Loss-per-Bag")
plt.text(0.1, 0.9, f"Correlation Cofficient: {round(np.corrcoef(X.reshape(-1),y.reshape(-1))[0][1],3)}",
         horizontalalignment='left', verticalalignment='center', transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', edgecolor='white', boxstyle='round,pad=0.5'))

plt.savefig('ppt_fig1.png',transparent=True)
```
